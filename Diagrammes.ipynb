```python
## 5 - Régression linéaire avec 2 caractéristiques et présentation des résultats
## dans un diagramme de dispersion, et avec une droite de régression.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("39-Versanddaten.csv", sep=";")
```

```python
plt.scatter(df["Produktkosten"], df["Gewicht"])
plt.xlabel("Produktkosten")
plt.ylabel("Gewicht")
plt.title("Diagramme de Dispersion")
```

```python
plt.scatter(df["Gewicht"], df["Produktkosten"])
plt.xlabel("Gewicht")
plt.ylabel("Produktkosten")
plt.title("Diagramme de Dispersion")
```

```python
import seaborn as sns
```

```python
sns.lmplot(x="Produktkosten", y="Gewicht",data=df)
```

```python
sns.lmplot(x="Gewicht", y="Produktkosten",data=df)
```
